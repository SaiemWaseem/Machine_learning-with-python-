# Support Vector Machine SVM 

### Import neccessary libraries

```python
import pandas as pd 
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

### Know Your Data

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
data.shape
```

### Visualize the similarities and differences of the species

```python
import warnings
warnings.filterwarnings('ignore')

sns.pairplot(data=data, hue='Species', palette= 'Paired')
```

### Pre-processing 


```python
le = preprocessing.LabelEncoder()
le.fit(data.Species)

data['Class_of_Species']= le.fit_transform(data.Species)
data.head()
```

### Tranform labels back to original encoding 

```python
le.inverse_transform(data['Class_of_Species'])
```

```python
X = data.iloc[:,:-2]
y = data.iloc[:,6]

X.head()
```

```python
y.head()
```

### Split Data into test and train sets

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = SVC(gamma='scale')

model.fit(x_train,y_train)
```

### - Prediction the SVC model

```python
prediction = model.predict(x_test)
```

### Evaluate the Model Performance

```python
print("Accuracy in Percentage:", round(metrics.accuracy_score(y_test, prediction),2)*100)
```
