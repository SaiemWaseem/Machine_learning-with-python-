### Clustering

```python
# import  necessary libraries 
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate sample data 
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)
```

```python
# instantiate a KMean object with number of cluster
kmeans = KMeans(n_clusters=4)
```

```python
# fit the data to the kmeans algorithm
kmeans.fit(X)
```

```python
# get the predicted cluster table for each plot
lables = kmeans.labels_
```

```python
# PLot the datapoint with different colors for cluster
plt.scatter(X[:, 0], X[:,1], c=lables)
plt.title("KMeans-Clustering")
plt.show()
```

### Clustering of Diamonds Dataset

```python
# import  necessary libraries 
from sklearn.cluster import KMeans
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('diamonds')
```

```python
data.head()
```

```python
# Convert categorial variables to numerical 
data['cut']= data['cut'].astype('category').cat.codes
data['color']= data['color'].astype('category').cat.codes
data['clarity']= data['clarity'].astype('category').cat.codes
```

```python
X = data[['carat','price','cut','color', 'clarity']]
```

```python
# instantiate a KMean object with number of cluster
kmeans = KMeans(n_clusters=5)
```

```python
# fit the data to the kmeans algorithm
kmeans.fit(X)
```

```python
# get the predicted cluster table for each plot
lables = kmeans.labels_
```

```python
# add the predicted cluster lable  as new column for original dataset
data['labels']= lables
```

```python
# PLot the datapoint with different colors for cluster
plt.scatter(X.iloc[:, 0], X.iloc[:,1], c=lables, cmap='viridis')
plt.xlabel('carat')
plt.ylabel('price')
plt.title("KMeans-Clustering on Diamond datas")
plt.show()
```

```python
# PLot the datapoint with different colors for cluster
plt.scatter(X.iloc[:, 0], X.iloc[:,2], c=lables, cmap='viridis')
plt.xlabel('carat')
plt.ylabel('color')
plt.title("KMeans-Clustering on Diamond datas")
plt.show()
```

```python
# PLot the datapoint with different colors for cluster
plt.scatter(X.iloc[:, 1], X.iloc[:,2], c=lables, cmap='viridis')
plt.xlabel('carat')
plt.ylabel('price')
plt.title("KMeans-Clustering on Diamond datas")
plt.show()
```
